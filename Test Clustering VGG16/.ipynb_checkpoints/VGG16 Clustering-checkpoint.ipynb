```python
from keras.preprocessing.image import load_img 
from keras.preprocessing.image import img_to_array 
from keras.applications.vgg16 import preprocess_input 

# models 
from keras.applications.vgg16 import VGG16 
from keras.models import Model

# clustering and dimension reduction
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

import os
import numpy as np
import matplotlib.pyplot as plt
from random import randint
import pandas as pd
import pickle
```

```python
path = r"D:\Ambis Kuliah\Semester 7\NAIST Internship\Breast Cancer Project\Program\YOLO crop train"

os.chdir(path)


mammography = []


with os.scandir(path) as files :
    for file in files :
        if file.name.endswith('.jpg'):
            mammography.append(file.name)
            
print(mammography[:15])
```

```python
img = load_img(mammography[0], target_size=(224,224))
img = np.array(img)

print(img.shape)
```

```python
reshaped_img = img.reshape(1,224,224,3)
print(reshaped_img.shape)
```

```python
x = preprocess_input(reshaped_img)
```

```python
model = VGG16()
model = Model(inputs=model.inputs, outputs=model.layers[-2].output)
```

```python
def extract_features(file, model):
    img = load_img(file, target_size=(224,224))
    img = np.array(img) 
    reshaped_img = img.reshape(1,224,224,3) 
    
    # prepare image for model
    imgx = preprocess_input(reshaped_img)
    
    # get the feature vector
    features = model.predict(imgx, use_multiprocessing=True)
    return features
```

```python
data={}

p = 'feat.pkl'

for mamph in mammography:
    try:
        feat = extract_features(mamph,model)
        data[mamph] = feat
    except:
        with open(p,'wb') as file:
            pickle.dump(data,file)
```

```python
filenames = np.array(list(data.keys()))

feat = np.array(list(data.values()))
feat.shape
```

```python
feat = feat.reshape(-1,4096)
feat.shape
```

```python
pca = PCA(n_components=100, random_state=22)
pca.fit(feat)
x = pca.transform(feat)


print(f"Components before PCA: {feat.shape[1]}")
print(f"Components after PCA: {pca.n_components}")
```

```python
kmeans = KMeans(n_clusters=9, random_state=22)
kmeans.fit(x)
```

```python
kmeans.labels_
```

```python
groups = {}
for file, cluster in zip(filenames,kmeans.labels_):
    if cluster not in groups.keys():
        groups[cluster] = []
        groups[cluster].append(file)
    else:
        groups[cluster].append(file)
```

```python
def view_cluster(cluster):
    plt.figure(figsize = (25,25));
    files = groups[cluster]
    if len(files) > 30:
        print(f"Clipping cluster size from {len(files)} to 30")
        files = files[:29]
    for index, file in enumerate(files):
        plt.subplot(10,10,index+1);
        img = load_img(file)
        img = np.array(img)
        plt.imshow(img)
        plt.axis('off')
```

```python
sse = []
list_k = list(range(3, 50))

for k in list_k:
    km = KMeans(n_clusters=k, random_state=22)
    km.fit(x)
    
    sse.append(km.inertia_)
```

```python
view_cluster(3)
```

```python
# Plot sse against k
plt.figure(figsize=(6, 6))
plt.plot(list_k, sse)
plt.xlabel(r'Number of clusters *k*')
plt.ylabel('Sum of squared distance');
```

```python
#elbow point otomatis
diff_wcss = np.diff(sse)
knee = np.argmax(diff_wcss) + 1

print("Elbow point (knee) at K =", knee)
```
